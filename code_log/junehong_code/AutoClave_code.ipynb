Auto clave 데이터 확인

## 1. 데이터 불러오기

```python
import warnings
import pandas as pd

# DtypeWarning 무시
warnings.filterwarnings('ignore', category=pd.errors.DtypeWarning)

# 데이터 파일 경로
file_path_csv = '../../data/Auto clave.csv'
file_path_excel = '../../data/Auto clave.xlsx'
file_path_target = '../../data/train_y.csv'

# 데이터 파일 읽기
csv_AutoClave = pd.read_csv(file_path_csv) # CSV 파일
excel_AutoClave = pd.read_excel(file_path_excel) # Excel 파일
csv_target = pd.read_csv(file_path_target) # CSV 파일

# csv_AutoClave & csv_target 병합
df_AutoClave = pd.merge(csv_AutoClave, csv_target, "inner", on="Set ID")
```

```python
# 출력 다 하고 싶을때 사용하는 코드

## 출력 모두 보기
pd.set_option('display.max_rows', None)

# -- 여기에 코드 작성 --

## 출력 모두 보기 취소 (원래대로 바꾸기)
pd.reset_option('display.max_rows')
```

```python
# csv 파일 정보 출력
csv_AutoClave.info()
```

```python
# excel 파일 정보 출력
excel_AutoClave.info()
```

```python
# 병합된 데이터 정보 출력
df_AutoClave.info()
```

```python
# 인덱스 0번째 행 출력 -> head로 출력하면 변수가 가로로 펴져서 보기 불편함
df_AutoClave.iloc[0, :]
```

```python
# excel 데이터의 각 열에 대한 값의 개수 출력 -> 변수명 확인용도
for col in excel_AutoClave.columns:
    counts = excel_AutoClave[col].value_counts()
    print(f"Value counts for column {col}:")
    print(counts)
    print()
```

```python
# 병합본 데이터의 각 열에 대한 값의 개수 출력 -> 실 사용용도
for col in df_AutoClave.columns:
    counts = csv_AutoClave[col].value_counts()
    print(f"Value counts for column {col}:")
    print(counts)
    print()
```
