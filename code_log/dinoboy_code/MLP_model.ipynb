```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('train_data_mlp.csv')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import backend as K
from sklearn.metrics import f1_score
import numpy as np

tf.random.set_seed(42)

#F1 스코어를 계산하는 사용자 정의 함수
def f1_m(y_true, y_pred):
    y_pred = K.round(y_pred)
    tp = K.sum(K.cast(y_true * y_pred, 'float'), axis=0)
    tn = K.sum(K.cast((1-y_true) * (1-y_pred), 'float'), axis=0)
    fp = K.sum(K.cast((1-y_true) * y_pred, 'float'), axis=0)
    fn = K.sum(K.cast(y_true * (1-y_pred), 'float'), axis=0)

    p = tp / (tp + fp + K.epsilon())
    r = tp / (tp + fn + K.epsilon())

    f1 = 2 * p * r / (p + r + K.epsilon())
    f1 = tf.where(tf.math.is_nan(f1), tf.zeros_like(f1), f1)
    return K.mean(f1)

# 모델 생성
model_dam = Sequential()
model_dam.add(Dense(40, input_dim=55, activation='relu'))
model_dam.add(Dense(25, activation='relu'))
model_dam.add(Dense(10, activation='relu'))
model_dam.add(Dense(1, activation='sigmoid'))

# F1 스코어를 평가지표로 사용
model_dam.compile(optimizer='adam', loss='binary_crossentropy', metrics=[f1_m])

model_dam.summary()
```

```python
tf.random.set_seed(12)
# 모델 생성
model_autoclave = Sequential()
model_autoclave.add(Dense(7, input_dim=9, activation='relu'))
model_autoclave.add(Dense(5, activation='relu'))
model_autoclave.add(Dense(3, activation='relu'))
model_autoclave.add(Dense(1, activation='sigmoid'))

# F1 스코어를 평가지표로 사용
model_autoclave.compile(optimizer='adam', loss='binary_crossentropy', metrics=[f1_m])

model_autoclave.summary()
```

```python
tf.random.set_seed(34)
# 모델 생성
model_fill1 = Sequential()
model_fill1.add(Dense(13, input_dim=19, activation='relu'))
model_fill1.add(Dense(8, activation='relu'))
model_fill1.add(Dense(4, activation='relu'))
model_fill1.add(Dense(1, activation='sigmoid'))

# F1 스코어를 평가지표로 사용
model_fill1.compile(optimizer='adam', loss='binary_crossentropy', metrics=[f1_m])

model_fill1.summary()
```

```python
tf.random.set_seed(25)
# 모델 생성
model_fill2 = Sequential()
model_fill2.add(Dense(13, input_dim=18, activation='relu'))
model_fill2.add(Dense(8, activation='relu'))
model_fill2.add(Dense(4, activation='relu'))
model_fill2.add(Dense(1, activation='sigmoid'))

# F1 스코어를 평가지표로 사용
model_fill2.compile(optimizer='adam', loss='binary_crossentropy', metrics=[f1_m])

model_fill2.summary()
```
